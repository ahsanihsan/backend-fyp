# FR : team a vs team b (who will win)

## yaml to csv set

```python
data = yaml.safe_load(open('./datasets/PKvsNZ_T20/_315.yml'))
len(data['innings'])
```

```python
directory = './datasets/'
outcome_of_matches = []
count=0
for j,sets in enumerate(os.listdir(directory)):
    if sets=='.DS_Store':
        continue
    current_directory =directory+sets+"/"
   
    for i,f in enumerate(os.listdir(current_directory)):
        print (current_directory+f)
        data = yaml.safe_load(open(current_directory+f))
        a=0
        score_a=0
        b=1
        score_b=0
        if len(data['innings']) != 2:
            continue
        if data['info']['teams'][0]==data['innings'][1]['2nd innings']['team']:
            a= 1
            b=0

        for d in data['innings'][a]['1st innings' if a==0 else '2nd innings']['deliveries']:
            score_a+=d[float(d.keys()[0])]['runs']['total']
        for d in data['innings'][b]['1st innings' if b==0 else '2nd innings']['deliveries']:
            score_b+=d[float(d.keys()[0])]['runs']['total']
    
        
        data=data['info']
        count+=1
        outcome_of_matches.append({
            'id': count,
            'city': data.get('city'),
            'month': ( datetime.strptime(data.get('dates')[0], "%Y-%m-%d") if type(data.get('dates')[0]) == str else data.get('dates')[0]).strftime("%B"),
            'match_type': data['match_type'],
            'team_a': data['teams'][0],
            'team_b': data['teams'][1],
            'toss_won': data['toss']['winner'],
            'toss_decision': data['toss']['decision'],
            'venue': data['venue'],
            'score_a':score_a,
            'score_b':score_b,
            'winner': data['outcome']['winner'] if 'winner' in data['outcome'] else 'draw'
        })
```

```python
score_b
```

```python
f= open('./csv/PKvsIND_ODI.csv','w+')
w = csv.DictWriter(f, ['id','city','month','match_type','team_a','team_b','toss_won','toss_decision','venue','score_a','score_b','winner'])
w.writeheader()
for obj in outcome_of_matches:
    w.writerow(obj)
f.close()
```

```python
import copy
directory = './datasets/PKvsIND_ODI/'
played_balls = []
id=0
for i,f in enumerate(os.listdir(directory)):
    print(f)
    load = yaml.safe_load(open(directory+f))
    data= load['info']
    obj={
        'id': id,
        'city': data.get('city'),
        'month': ( datetime.strptime(data.get('dates')[0], "%Y-%m-%d") if type(data.get('dates')[0]) == str else data.get('dates')[0]).strftime("%B"),
        'match_type': data['match_type'],
        'venue': data['venue'],
        'batsman':'',
        'bowler':'',
        'over_no':'',
        'ball_no':'',
        'wicket_out':'',
        'score':'',
        'position':'',
        'team':''
        
    }
    for innings in load['innings']:
       
        first=innings[list(innings.keys())[0]]
        obj['team'] = first['team']
        first = first['deliveries'][0][0.1]
        p=3
        positions = {first['batsman']:1,first['non_striker']:2}
        for d in innings[list(innings.keys())[0]]['deliveries']:
            ball = str(list(d.keys())[0])
            over_ball=ball.split('.')
            obj['over_no']=over_ball[0]
            id+=1
            obj['wicket_out']=False
            obj['id']=id
            obj['ball_no']=over_ball[1]
            obj['bowler']=d[float(ball)]['bowler']
            obj['batsman']=d[float(ball)]['batsman']
            if obj['batsman'] in positions:
                obj['position']=positions[obj['batsman']]
            else:
                obj['position']=p
                positions[obj['batsman']]=p
                p+=1
                
            obj['score']=d[float(ball)]['runs']['total']
            if 'wicket' in d[float(ball)]:
                if obj['score']==0:
                    if obj['batsman']==d[float(ball)]['wicket']['player_out']:
                        obj['wicket_out']=True
                    else:
                        obj['wicket_out']=False
                        played_balls.append(obj)
                        obj['batsman']=d[float(ball)]['wicket']['player_out']
                        obj['score']=0

            else:
                d[float(ball)]['wicket_out'] = False
            played_balls.append(copy.deepcopy(obj))
```

```python
f= open('./csv/PKvsIND_ODI_players.csv','w+')
w = csv.DictWriter(f, ['id','city','month','match_type','team','batsman','position','bowler','over_no','ball_no','venue','wicket_out','score'])
w.writeheader()
for obj in played_balls:
    w.writerow(obj)
f.close()
```

# FR: Runrate prediction 

```python
directory = './datasets/'
outcome_of_matches = []
id=0
for dir in os.listdir(directory):
    if dir==".DS_Store":
        continue
    curr_dir = directory+dir+"/"
    for j,f in enumerate(os.listdir(curr_dir)):
        data = yaml.safe_load(open(curr_dir+f))
        obj={
            'id':0,
            'batting_team':data['innings'][0]['1st innings']['team'],
            'bowling_team':data['info']['teams'][0] if data['info']['teams'][1] ==data['innings'][0]['1st innings']['team'] else data['info']['teams'][1],
            'run_rate':0,
            'match_type':data['info']['match_type']
        }
        for innings in data['innings']:
            runs=0
            obj['batting_team']=innings[list(innings.keys())[0]]['team']
            obj['bowling_team']=data['info']['teams'][0] if data['info']['teams'][1] ==innings[list(innings.keys())[0]]['team'] else data['info']['teams'][1]
            over=0
            for d in innings[list(innings.keys())[0]]['deliveries']:
    #             if str(obj['over_no']) == str(d.keys()[0]).split(".")[0]:
    #                 runs+=d[list(d.keys())[0]]['runs']['total']

    #             else:
    #                 obj['run_rate']=runs/6
    #                 outcome_of_matches.append(copy.deepcopy(obj))
    #                 runs=d[list(d.keys())[0]]['runs']['total']
    #                 obj['over_no'] = int(str(d.keys()[0]).split(".")[0])
                runs+=d[list(d.keys())[0]]['runs']['total']
                over=int(str(d.keys()[0]).split(".")[0])
            obj['run_rate']=runs/(over+1)
            id+=1
            obj['id']=id
            outcome_of_matches.append(copy.deepcopy(obj))
```

```python
f= open('./csv/runrates.csv','w+')
w = csv.DictWriter(f, ['id','match_type','batting_team','bowling_team','run_rate'])
w.writeheader()
for obj in outcome_of_matches:
    w.writerow(obj)
f.close()
```
